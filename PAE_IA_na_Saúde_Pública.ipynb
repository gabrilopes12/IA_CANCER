````python
!gdown 1KH88q3hSuoAQHDo10vLBZjvQaI0U3gID --quiet

df = pd.read_csv('/content/colorretal_3anos.csv')
print(df.shape)
df.head(3)
````

````python
X = df.drop(['ECGRUP', 'ULTINFO', 'ULTIDIAG', 'obito_geral', 'sobrevida_ano3'],
            axis=1)
y = df.sobrevida_ano3
````

````python
from sklearn.model_selection import train_test_split

X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.25,
                                                    random_state=5)

X_train.shape, X_test.shape, y_train.shape, y_test.shape
````

````python
from sklearn.preprocessing import OrdinalEncoder

oe = OrdinalEncoder(handle_unknown='use_encoded_value',
                    unknown_value=-1)
oe.fit(X_train[['EC']])
````

````python
X_train[['EC']] = oe.transform(X_train[['EC']])
X_train[['EC']]
````

# **One Hot Encoder**

````python
from sklearn.preprocessing import OneHotEncoder

X_train_ohe = X_train.copy()

ohe = OneHotEncoder(handle_unknown='ignore', drop='first')

ohe.fit(X_train_ohe[['TOPO']])
````

````python
ohe_results = ohe.transform(X_train_ohe[['TOPO']]).toarray()
ohe_results[:4]
````

````python
df_aux = pd.DataFrame(ohe_results, index=X_train_ohe.index,
                      columns=ohe.get_feature_names_out())
df_aux.head(3)
````

````python
X_train_ohe.head(3)
````

````python
X_train_ohe = X_train_ohe.merge(df_aux, how='left', left_index=True,
                                right_index=True)

X_train_ohe.head(3)
````

````python
X_train_ohe.TOPO.head(3)
````

````python
X_train_ohe.drop(columns=['TOPO'], inplace=True)
X_train_ohe.columns
````

````python
X_train_ohe.select_dtypes(include='object').columns
````

# **Normalização**

````python
from sklearn.preprocessing import MinMaxScaler

mms = MinMaxScaler()
mms.fit(X_train_ohe)
````

````python
X_train_norm = mms.transform(X_train_ohe)
X_train_norm
````

````python
X_train_norm.min()
````

````python
X_train_norm.max()
````

````python
X_test
````

# **Transformar os dados de Teste**

````python
# X_test
X_test_enc = X_test.copy()

# Ordinal Encoder
X_test_enc[['EC']] = oe.transform(X_test_enc[['EC']])

# One Hot Encoder
ohe_results = ohe.transform(X_test_enc[['TOPO']]).toarray()
df_aux = pd.DataFrame(ohe_results, index=X_test_enc.index,
                      columns=ohe.get_feature_names_out())
X_test_enc = X_test_enc.merge(df_aux, how='left', left_index=True,
                              right_index=True)
X_test_enc.drop(columns=['TOPO'], inplace=True)

# Normalização
X_test_norm = mms.transform(X_test_enc)
````

````python
X_test_norm
````

````python
X_test_norm.shape, X_train_norm.shape
````

# **Random Forest**

## **Modelo Baseline**

````python
from sklearn.ensemble import RandomForestClassifier

rf_base = RandomForestClassifier(random_state=5, max_depth=8)
rf_base.fit(X_train_norm, y_train)
````

````python
rf_base.score(X_test_norm, y_test)
````

````python
y_train.value_counts(normalize=True)
````

````python
from sklearn.metrics import ConfusionMatrixDisplay
import matplotlib.pyplot as plt

y_pred = rf_base.predict(X_test_norm)
ConfusionMatrixDisplay.from_predictions(y_test, y_pred, normalize='true',
                                        values_format='.4f')
plt.show()
````

````python
from sklearn.metrics import roc_curve, auc

model = rf_base

# Calcula as probabilidades de pertencer à classe positiva para os dados de treinamento e teste usando o modelo.
probas_train = model.predict_proba(X_train_norm)[:, 1]
probas_test = model.predict_proba(X_test_norm)[:, 1]

# Calcula a taxa de falsos positivos (FP) e a taxa de verdadeiros positivos (TP) para os dados de treinamento e teste.
fp_train, tp_train, _ = roc_curve(y_train, probas_train)
fp_test, tp_test, _ = roc_curve(y_test, probas_test)

# Cria uma figura com o tamanho especificado.
plt.figure(figsize=(10, 7))

# Plota a curva ROC para os dados de treinamento com a taxa de falsos positivos no eixo x e a taxa de verdadeiros positivos no eixo y, utilizando a cor azul.
plt.plot(fp_train, tp_train, 'b', label=f'Train (AUC = {auc(fp_train, tp_train):.4f})')

# Plota a curva ROC para os dados de teste com a taxa de falsos positivos no eixo x e a taxa de verdadeiros positivos no eixo y, utilizando a cor vermelha.
plt.plot(fp_test, tp_test, 'r', label=f'Test (AUC = {auc(fp_test, tp_test):.4f})')

# Plota a linha de base (diagonal) como referência, utilizando linha pontilhada preta.
plt.plot(np.linspace(0, 1, 100),
         np.linspace(0, 1, 100),
         label='Baseline',
         linestyle='--',
         color='k')

# Adiciona rótulos aos eixos x e y.
plt.xlabel('False Positives')
plt.ylabel('True Positives')

# Adiciona uma grade ao gráfico.
plt.grid(True)

# Adiciona uma legenda ao gráfico.
plt.legend()

# Exibe o gráfico.
plt.show()
````
